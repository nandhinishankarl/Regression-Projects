## Loading the diabetes dataset 

### Import libary 

```python
from sklearn import datasets 
```

### Load dataset

```python
diabetes = datasets.load_diabetes()
```

```python
diabetes
```

### Description of the dataset

```python
print(diabetes.DESCR)
```

### Feature names

```python
print(diabetes.feature_names)
```

```python
print(diabetes.target)
```

### Create the X and Y data matrices 

```python
X = diabetes.data
Y = diabetes.target
```

```python
X.shape, Y.shape 
```

### Data split

```python
from sklearn.model_selection import train_test_split
```

### 80-20 data split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

### Data dimensions

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

### Linear regression model 

```python
# allows you to build the linear regression model
from sklearn import linear_model
# allows you to compute the performance of the model 
from sklearn.metrics import mean_squared_error, r2_score
```

### Build linear regression model 

```python
model = linear_model.LinearRegression()
```

### Build training model 

```python
# we're using 80% of the data as the training set to build the model 
model.fit(X_train, Y_train)
```

```python
# assigning the built model in order to make the prediction
# on 20% of the dataset 
Y_pred = model.predict(X_test)
```

### Prediction results 

```python
# the coefficients are for the features 
print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
print('Mean squared error (MSE): %.2f'
      % mean_squared_error(Y_test, Y_pred))
print('Coefficient of determination (R^2): %.2f'
      % r2_score(Y_test, Y_pred))
```
